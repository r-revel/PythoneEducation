## Задание 1: Работа с массивами NumPy


Задание 1: Работа с массивами NumPy

**Цель**: Проверить знание методов NumPy для анализа данных.

**Задание**:

1.  Создайте массив случайных чисел (50 элементов) из диапазона [10, 100].
2.  Найдите:
3.  Среднее арифметическое массива.
4.  Медиану.
5.  Стандартное отклонение.
6.  Создайте новый массив, где каждое число изначального массива делится на 3. Отфильтруйте числа, которые делятся без остатка.

**Ожидаемый ответ**:

-   Вывод среднего, медианы, стандартного отклонения.
-   Новый массив и его фильтр.

**Балл**: 6

```python
# Шаг 1: Создание массива случайных чисел
np.random.seed(42)  # Для воспроизводимости результатов
random_array = np.random.randint(10, 101, size=50)

# Шаг 2: Вычисление среднего, медианы и стандартного отклонения
mean_value = np.mean(random_array)
median_value = np.median(random_array)
std_deviation = np.std(random_array)

# Вывод результатов
print(f"Среднее арифметическое: {mean_value}")
print(f"Медиана: {median_value}")
print(f"Стандартное отклонение: {std_deviation}")

# Шаг 3: Создание нового массива и фильтрация
divided_array = random_array / 3
filtered_array = random_array[random_array % 3 == 0]

# Вывод нового массива и его фильтра
print(f"Новый массив (деление на 3): {divided_array}")
print(f"Фильтрованный массив (делится без остатка): {filtered_array}")
```

## Задание 3: Очистка данных и визуализация


**Цель**: Проверить умение находить и исправлять ошибки в данных, а также визуализировать их.

**Данные:**  Используйте предоставленный CSV-файл ([sales_cleaning_large.csv](https://lms.skillfactory.ru/asset-v1:SkillFactory+MIFIML-1sem+2024+type@asset+block@sales_cleaning_large.csv)), содержащий данные о продажах с возможными пропущенными и некорректными значениями. Поля:

-   `date`  (дата продаж),
-   `sales`  (объем продаж),
-   `region`  (регион).

**Задание**:

1.  Найдите и устраните ошибки в данных:
2.  Заполните пропущенные даты.
3.  Удалите отрицательные значения в столбце sales или замените их на медиану столбца.
4.  Заполните пропущенные значения в sales средним значением.
5.  Постройте график продаж по датам. Убедитесь, что даты упорядочены.

**Ожидаемый ответ**:

-   Обработанный DataFrame.
-   Построенный график.

```python
# Загрузка данных
df = pd.read_csv('sales_cleaning_large.csv', parse_dates=['date'])

# Проверка первых строк данных
print(df.head())

# 1. Преобразование столбца 'date' в формат даты (если не было преобразовано автоматически)
df['date'] = pd.to_datetime(df['date'])

# Проверяем на наличие дубликатов
print('Дубликатов: ', df.duplicated().sum())

# Заполнение пропущенных дат
last_known_date = df['date'].dropna().max()
next_date = pd.to_datetime(last_known_date) + pd.Timedelta(days=1)
df['date'] = df['date'].fillna(next_date.strftime('%Y-%m-%d'))

# 3. Заполнение пропущенных дат и регионов
# Создаем диапазон дат от минимальной до максимальной
date_range = pd.date_range(start=df['date'].min(), end=df['date'].max(), freq='D')
regions = df['region'].unique()

# Удаляем дубликаты по столбцам date и region, оставляя первое вхождение
df = df.drop_duplicates(subset=['date', 'region'], keep='first')

# Создаем DataFrame с полным набором дат и регионов
full_df = pd.DataFrame({
    'date': date_range.repeat(len(regions)),
    'region': regions.tolist() * len(date_range)
})

# Объединяем полный DataFrame с оригинальными данными
df = pd.merge(full_df, df, on=['date', 'region'], how='left')

# 4. Удаление отрицательных значений в столбце sales или замена на медиану
median_sales = df['sales'][df['sales'] > 0].median()
df['sales'] = df['sales'].apply(lambda x: median_sales if x < 0 else x)

# 5. Заполнение пропущенных значений в sales с использованием интерполяции
df['sales'] = df.groupby('region')['sales'].transform(lambda x: x.interpolate(method='linear'))

# Дополнительное заполнение пропущенных значений в начале и конце временного ряда
df['sales'] = df.groupby('region')['sales'].transform(lambda x: x.fillna(median_sales))

# Проверка на наличие дубликатов
print('Дубликатов после обработки: ', df.duplicated().sum())

# 6. Построение стекированного графика продаж по датам
plt.figure(figsize=(10, 6))

# Подготовка данных для stackplot
sales_by_region = [df[df['region'] == region]['sales'].values for region in regions]

plt.stackplot(date_range, sales_by_region, labels=regions)

plt.title('Продажи по датам и регионам')
plt.xlabel('Дата')
plt.ylabel('Объем продаж')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()

# Проверка обработанного DataFrame
print(df.head())
print(df.tail())
```

## Задание 4: Получение данных из API

**Цель**: Проверить способность работать с веб-данными.

**Задание**:

1.  Используйте открытое API (например,  [https://jsonplaceholder.typicode.com/users](https://jsonplaceholder.typicode.com/users)).
2.  Загрузите данные и создайте DataFrame с полями:  `id`,  `name`,  `email`,  `company_name`.
3.  Выведите таблицу с пользователями, работающими в компаниях, название которых содержит слово "Group".

**Ожидаемый ответ**:

-   DataFrame с результатами фильтрации.

```python

# Шаг 2: Загрузка данных с API
url = 'https://jsonplaceholder.typicode.com/users'
response = requests.get(url)

# Проверка успешности запроса
if response.status_code == 200:
    data = response.json()
else:
    print(f"Ошибка при загрузке данных: {response.status_code}")
    data = []

# Шаг 3: Создание DataFrame
user_data = [
    {
        'id': user['id'],
        'name': user['name'],
        'email': user['email'],
        'company_name': user['company']['name']
    }
    for user in data
]

df = pd.DataFrame(user_data)

# Шаг 4: Фильтрация DataFrame
filtered_df = df[df['company_name'].str.contains('Group', case=False)]

# Шаг 5: Вывод результата
filtered_df
```
