
Задание 1: Работа с массивами NumPy

**Цель**: Проверить знание методов NumPy для анализа данных.

**Задание**:

1.  Создайте массив случайных чисел (50 элементов) из диапазона [10, 100].
2.  Найдите:
3.  Среднее арифметическое массива.
4.  Медиану.
5.  Стандартное отклонение.
6.  Создайте новый массив, где каждое число изначального массива делится на 3. Отфильтруйте числа, которые делятся без остатка.

**Ожидаемый ответ**:

-   Вывод среднего, медианы, стандартного отклонения.
-   Новый массив и его фильтр.

**Балл**: 6

```python
import numpy as np

# Шаг 1: Создание массива случайных чисел
np.random.seed(42)  # Для воспроизводимости результатов
random_array = np.random.randint(10, 101, size=50)

# Шаг 2: Вычисление среднего, медианы и стандартного отклонения
mean_value = np.mean(random_array)
median_value = np.median(random_array)
std_deviation = np.std(random_array)

# Вывод результатов
print(f"Среднее арифметическое: {mean_value}")
print(f"Медиана: {median_value}")
print(f"Стандартное отклонение: {std_deviation}")

# Шаг 3: Создание нового массива и фильтрация
divided_array = random_array / 3
filtered_array = random_array[random_array % 3 == 0]

# Вывод нового массива и его фильтра
print(f"Новый массив (деление на 3): {divided_array}")
print(f"Фильтрованный массив (делится без остатка): {filtered_array}")
```
