Обратный порядок слов в блоках текста

Дан текстовый файл, каждое предложение которого занимает одну строку. Напишите программу, которая разделяет текст на блоки — каждый блок состоит из нескольких предложений. Затем переворачивает порядок слов только внутри каждого предложения, не меняя порядок самих предложений в блоке.

Файл содержит:

    Привет как дела
    На улице идет дождь
    Я люблю программирование

Критерии оценки:

    Текст разделен на блоки корректно — 2 балла.
    Корректно перевернуты блоки — 2 балла.


```python
with open('task1_file1.txt', 'w', encoding='utf-8') as f:
    f.write('Привет как дела\nНа улице идет дождь\nЯ люблю программирование')
with open('task1_file2.txt', 'w', encoding='utf-8') as f:
    f.write('Сегодня погода отличная\nУ нас солнечно\nЗавтра ожидается дождь\nПослезавтра снова будет солнечно')
with open('task1_file3.txt', 'w', encoding='utf-8') as f:
    f.write('Один\nДва\nТри\nЧетыре\nПять\nШесть\nСемь')
```

```python
def reverse_words_in_blocks(text):
    # Разделяем текст на строки (предложения)
    lines = text.split('\n')
    # Инициализируем результат
    result = []
    # Обрабатываем каждую строку как отдельный блок
    for line in lines:
        # Переворачиваем порядок слов в предложении
        reversed_words = ' '.join(line.split()[::-1])
        result.append(reversed_words)
    # Возвращаем результат в виде строки
    return '\n'.join(result)


# Список файлов для обработки
files = ['task1_file1.txt', 'task1_file2.txt', 'task1_file3.txt']

# Обрабатываем каждый файл
for file_name in files:
    try:
        with open(file_name, 'r', encoding='utf-8') as f:
            text = f.read()
            reversed_text = reverse_words_in_blocks(text)
            print(f"Содержимое файла {file_name} после переворота слов:")
            print(reversed_text)
            print('-' * 40)
    except FileNotFoundError:
        print(f"Файл {file_name} не найден.")
    except Exception as e:
        print(f"Произошла ошибка при обработке файла {file_name}: {e}")
```

Задание 2 (2 балла)

Напишите функцию, которая принимает строку и сжимает её определённым образом. Строки содержат произвольные символы, включая пробелы и спецсимволы, и требуют точного учета длины при кодировании. Сжатие строки происходит сериями одинаковых символов в формате символколичество, но только если длина сжатой строки не превышает исходную.

Примеры:

    вход: aaabbc → вывод: a3b2c,
    вход: abcd → вывод: abcd (так как сжатая равна).

Критерии оценки:

    Корректно очищен текст — 1 балл.
    Выполняется условие — 1 балл.


```python
def compress_string(s):
    if not s:
        return s 
    compressed = []
    # Счетчик для подсчета одинаковых символов
    count = 1
    for i in range(1, len(s)):
        # Если текущий символ совпадает с предыдущим, увеличиваем счетчик
        if s[i] == s[i - 1]:
            count += 1
        else:
            # Если символы не совпадают, добавляем предыдущий символ и его количество (если больше 1)
            compressed.append(f"{s[i - 1]}{count if count > 1 else ''}")
            # Сбрасываем счетчик для нового символа
            count = 1
    # Добавляем последний символ и его количество
    compressed.append(f"{s[-1]}{count if count > 1 else ''}")
    # Объединяем все части в одну строку
    compressed_str = ''.join(compressed)
    # Возвращаем сжатую строку, если она короче или равна исходной, иначе возвращаем исходную строку
    return compressed_str if len(compressed_str) <= len(s) else s

def test_compress_string():
    test_cases = [
        ("aaabbc", "a3b2c"),
        ("abcd", "abcd"),
        ("a b c d", "a b c d"),
        ("!!!###$", "!3#3$"),
        ("aaaabbbbbccccccc", "a4b5c7"),
        ("", ""),
        ("a", "a"),
        ("aa bb cc dd", "a2 b2 c2 d2"),
        ("abcdefghijklmnopqrstuvwxyz", "abcdefghijklmnopqrstuvwxyz"),
        ("aabbcc", "a2b2c2"),
        ("abc", "a1b1c1")
    ]

    for input_str, expected_output in test_cases:
        assert compress_string(input_str) == expected_output, f"Ошибка данные: {input_str}, результат: {compress_string(input_str)}"
        print(f"Пройдено для: {input_str} -> {expected_output}")

test_compress_string()
```

Задание 3 (4 балла)
Хаотичные скобки

Реализуйте функцию, которая проверяет, правильно ли расставлены скобки в строке (включая круглые, квадратные и фигурные скобки).

Критерии оценки:

    Функция выполняет условие — 4 балла.


```python
def is_balanced_brackets(s):
    bracket_map = {')': '(', ']': '[', '}': '{'}
    stack = []

    for char in s:
        # Если символ — открывающая скобка, добавляем её в стек
        if char in bracket_map.values():
            stack.append(char)
        # Если символ — закрывающая скобка
        elif char in bracket_map:
            # Если стек пуст или верхний элемент стека не соответствует открывающей скобке, возвращаем False
            if not stack or stack.pop() != bracket_map[char]:
                return False
        # Игнорируем другие символы
        else:
            continue

    # Если стек пуст, все скобки правильно закрыты
    return not stack

def test_is_balanced_brackets():
    test_cases = [
        ("()", True),
        ("[]", True),
        ("{}", True),
        ("(())", True),
        ("[[]]", True),
        ("{{}}", True),
        ("(())[]{}", True),
        ("([{}])", True),
        ("(]", False),
        ("([)]", False),
        ("{[)]}", False),
        ("{[}", False),
        ("{[()]}[]", True),
        ("", True),
        ("a(b[c]d)e", True),
        ("a(b[c)d)e", False)
    ]

    for input_str, expected_output in test_cases:
        assert is_balanced_brackets(input_str) == expected_output, f"Ошибка данные: {input_str}, результат: {is_balanced_brackets(input_str)}"
        print(f"Пройдено для: {input_str} -> {expected_output}")

test_is_balanced_brackets()
```

Задание 4 (4 балла)
Генератор случайных паролей

Напишите функцию, которая генерирует пароль заданной длины. В реализации надо учитывать, что:

    Пароль должен содержать буквы, цифры и специальные символы.
    Длина пароля задается пользователем.

Критерии оценки:

    Функция выполняет все заявленные условия — 4 балла.


```python

def generate_password(length):
    if length < 6:
        raise ValueError("Длина пароля должна быть не менее 6 символов.")
    # Определяем наборы символов
    lowercase_letters = string.ascii_lowercase  # строчные буквы
    uppercase_letters = string.ascii_uppercase  # заглавные буквы
    digits = string.digits  # цифры
    special_characters = string.punctuation  # специальные символы
    # Объединяем все символы в один список
    all_characters = lowercase_letters + uppercase_letters + digits + special_characters
    # Убедимся, что пароль содержит хотя бы по одному символу каждого типа
    password = [
        random.choice(lowercase_letters),
        random.choice(uppercase_letters),
        random.choice(digits),
        random.choice(special_characters)
    ]
    # Добавляем оставшиеся символы до достижения заданной длины
    password += random.choices(all_characters, k=length - 4)
    # Перемешиваем символы для повышения случайности
    random.shuffle(password)
    # Преобразуем список в строку
    return ''.join(password)

def test_generate_password():
    # Тестирование с разными длинами пароля
    test_lengths = [6, 8, 12, 16, 20]
    for length in test_lengths:
        password = generate_password(length)
        print(f"Пароль длиной {length} символов: {password}")
        assert len(password) == length, f"Длина пароля {length} не соответствует ожидаемой."
        assert any(c in string.ascii_lowercase for c in password), "Пароль не содержит строчных букв."
        assert any(c in string.ascii_uppercase for c in password), "Пароль не содержит заглавных букв."
        assert any(c in string.digits for c in password), "Пароль не содержит цифр."
        assert any(c in string.punctuation for c in password), "Пароль не содержит специальных символов."

test_generate_password()
```

Задание 5 (6 баллов)
Эмуляция работы электронной очереди

Напишите класс для симуляции работы электронной очереди, например, в банке.

Критерии оценки:

    Система должна поддерживать добавление клиентов с указанием их приоритета (например, VIP, обычный) — 1 балл.
    Выбор клиента для следующей обработки должен учитывать приоритет — 1 балл.
    Реализуйте отчет для администрации с информацией, сколько времени заняло обслуживание клиентов — 4 балла.


```python
class Client:
    def __init__(self, name, priority, service_time):
        self.name = name
        self.priority = priority
        self.service_time = service_time
        self.start_time = None
        self.end_time = None

    def __lt__(self, other):
        # Определяем порядок для кучи (меньший приоритет — больший приоритет)
        return self.priority < other.priority

    def start_service(self):
        self.start_time = time.time()

    def end_service(self):
        self.end_time = time.time()

    def get_service_duration(self):
        return self.end_time - self.start_time

class PriorityQueue:
    def __init__(self):
        self.queue = []

    def add_client(self, client):
        heapq.heappush(self.queue, client)

    def get_next_client(self):
        if self.queue:
            return heapq.heappop(self.queue)
        return None

    def is_empty(self):
        return len(self.queue) == 0

class QueueSystem:
    def __init__(self):
        self.queue = PriorityQueue()
        self.report = []

    def add_client(self, name, priority, service_time):
        client = Client(name, priority, service_time)
        self.queue.add_client(client)

    def process_next_client(self):
        client = self.queue.get_next_client()
        if client:
            client.start_service()
            time.sleep(client.service_time)  # Эмуляция времени обслуживания
            client.end_service()
            self.report.append(client)
            print(f"Обслужен клиент: {client.name}, время обслуживания: {client.get_service_duration():.2f} секунд")
        else:
            print("Очередь пуста.")

    def generate_report(self):
        total_service_time = 0
        for client in self.report:
            total_service_time += client.get_service_duration()
            print(f"Клиент: {client.name}, время обслуживания: {client.get_service_duration():.2f} секунд")
        print(f"Общее время обслуживания клиентов: {total_service_time:.2f} секунд")

# Пример использования
if __name__ == "__main__":
    queue_system = QueueSystem()
    queue_system.add_client("Анна", 1, 5)  # VIP клиент
    queue_system.add_client("Иван", 2, 3)  # Обычный клиент
    queue_system.add_client("Мария", 1, 4)  # VIP клиент
    queue_system.add_client("Петр", 2, 2)  # Обычный клиент

    # Обработка клиентов
    while not queue_system.queue.is_empty():
        queue_system.process_next_client()
    print('\n##Генерация отчета##\n')
    queue_system.generate_report()
```

Задание 6 (2 балла)
Проверка на «почти палиндром»

Напишите программу, которая проверяет, является ли строка палиндромом или «почти палиндромом». «Почти палиндром» означает, что можно удалить одну букву, чтобы строка стала палиндромом.

Критерии оценки:

    Функция выполняет все заявленные условия — 2 балла.


```python
def is_palindrome(s):
    # Удаляем все специальные символы и пробелы, приводим к нижнему регистру
    cleaned_s = re.sub(r'[^a-zA-Z0-9]', '', s).lower()
    return cleaned_s == cleaned_s[::-1]

def is_almost_palindrome(s):
    # Удаляем все специальные символы и пробелы, приводим к нижнему регистру
    cleaned_s = re.sub(r'[^a-zA-Z0-9]', '', s).lower()
    
    # Проверяем, является ли строка палиндромом
    if is_palindrome(cleaned_s):
        return "Палиндром"
    
    # Проверяем, можно ли удалить одну букву, чтобы строка стала палиндромом
    for i in range(len(cleaned_s)):
        # Удаляем одну букву и проверяем, является ли оставшаяся строка палиндромом
        if is_palindrome(cleaned_s[:i] + cleaned_s[i+1:]):
            return "Почти палиндром"
    
    return "Не палиндром и не почти палиндром"

def test_is_almost_palindrome():
    test_cases = [
        ("racecar", "Палиндром"),  
        ("radar", "Палиндром"),    
        ("abcba", "Палиндром"),    
        ("abccba", "Палиндром"),   
        ("abccaa", "Не палиндром и не почти палиндром"),  
        ("abccan", "Не палиндром и не почти палиндром"),  
        ("abcbaa", "Почти палиндром"),   
        ("abccaa", "Не палиндром и не почти палиндром"), 
        ("deified", "Палиндром"),
        ("a", "Палиндром"),      
        ("aa", "Палиндром"),   
        ("ab", "Почти палиндром"),     
        ("abc", "Не палиндром и не почти палиндром"),  
        ("A man a plan a canal Panama", "Палиндром"), 
        ("Was it a car or a cat I saw", "Палиндром"), 
        ("Madam as Eden, I’m Adam", "Не палиндром и не почти палиндром"),  
        ("A Toyota", "Палиндром"),
        ("No lemon, no melon", "Палиндром")
    ]

    for input_str, expected_output in test_cases:
        result = is_almost_palindrome(input_str)
        assert result == expected_output, f"Ошибка данные: {input_str}, результат: {result} ожидаю: {expected_output}"
        print(f"Input: {input_str} -> {result}")

test_is_almost_palindrome()
```

Задание 7 (9 баллов)
Задача Шредингера
Разработайте программу, которая «стирает» фрагменты текста в файле. Например:

Пользователь указывает файл и процент текста, который нужно удалить (например, 30%).
Программа случайно выбирает слова или части абзацев и заменяет их на пробел или ..., сохраняя общий объем документа.
Вход:
«Сегодня солнечный день, и я собираюсь гулять в парке с моими друзьями.»

Вывод (удалено ~30%):
«Сегодня ... день, и я собираюсь гулять ... моими друзьями.»

Критерии оценки:

    Чтение файла и ввод параметров пользователя — 3 балла.
    Реализация алгоритма случайного удаления текста — 5 баллов.
    Запись результата и вывод пользователю — 1 балл.

```python


def erase_text(text, percent_to_erase, fill_char='...'):
    # Разбиваем текст на слова и создаем словарь с количеством символов в каждом слове
    words = text.split()
    word_lengths = {word: len(word) for word in words}
    
    # Общая длина текста
    total_length = sum(word_lengths.values())
    
    # Определяем количество символов, которое нужно удалить
    target_length_to_erase = int(total_length * (percent_to_erase / 100.0))
    
    # Найдем оптимальную комбинацию слов для удаления
    def find_optimal_combination(word_lengths, target_length):
        best_combination = []
        best_diff = float('inf')
        
        for r in range(1, len(word_lengths) + 1):
            # нашел готовую функцию для создания всех комбинаций можно было конечно сделать с numpy или руками но так показалось лучше всего
            for combo in combinations(word_lengths.items(), r):
                combo_length = sum(length for word, length in combo)
                diff = abs(combo_length - target_length)
                
                if diff < best_diff:
                    best_diff = diff
                    best_combination = [word for word, length in combo]
        
        return best_combination
    
    words_to_erase = find_optimal_combination(word_lengths, target_length_to_erase)
    
    # Заменяем выбранные слова на fill_char
    for index, word in enumerate(words):
        if word in words_to_erase:
            words[index] = fill_char
    
    # Собираем текст обратно
    erased_text = ' '.join(words)
    
    return erased_text

def calculate_actual_percent_erased(original_text, erased_text, fill_char='...'):
    original_length = len(original_text)
    erased_length = len(erased_text.replace(fill_char, ''))
    actual_percent_erased = ((original_length - erased_length) / original_length) * 100
    return actual_percent_erased

def main():
    examples = [
        ("Сегодня солнечный день, и я собираюсь гулять в парке с моими друзьями.", 30),
        ("Это пример текста, который будет обработан программой.", 50),
        ("Python — это мощный язык программирования, который используется в различных областях.", 20),
        ("Программа должна учитывать регистр и специальные символы.", 99)
    ]
    fill_char = '...'  # Тип заполнения

    for example, percent in examples:
        erased_text = erase_text(example, percent, fill_char)
        actual_percent_erased = calculate_actual_percent_erased(example, erased_text, fill_char)
        print(f"Оригинальный текст: {example}")
        print(f"Удалено ~{percent}%: {erased_text}")
        print(f"Фактический процент удаленных символов: {actual_percent_erased:.2f}%")
        print('-' * 80)

if __name__ == "__main__":
    main()
```
