## Задание 11

Сколькими способами можно поставить 5 различных книг на полке, если две из них должны стоять рядом?

У нас осталось 3 книги и 1 блок (состоящий из двух книг), то есть всего 4 элемента. Количество перестановок 4 элементов равно $4!$:

$
4! = 4 \times 3 \times 2 \times 1 = 24
$

Внутри блока могут стоять в двух возможных порядках. Количество перестановок 2 элементов равно $2!$:

$

$
4! \times 2! = 24 \times 2 = 48
$

Ответ:

$ 48 $

## Задание 12

Вероятность, что первый телефон сломан - 0.1. Вероятность того, что второй телефон сломан - 0.3. Найдите вероятность, что оба телефона работают.

Вероятность, что первый телефон сломан, равна $0.1$. Следовательно, вероятность, что первый телефон работает, равна:
   $
   P(\text{первый телефон работает}) = 1 - 0.1 = 0.9
   $

Вероятность, что второй телефон сломан, равна $0.3$. Следовательно, вероятность, что второй телефон работает, равна:
   $
   P(\text{второй телефон работает}) = 1 - 0.3 = 0.7
   $

Поскольку события независимы, вероятность, что оба телефона работают, равна произведению вероятностей того, что каждый из них работает:

   $
   P(\text{оба телефона работают}) = P(\text{первый телефон работает}) \times P(\text{второй телефон работает}) = 0.9 \times 0.7 = 0.63
   $

Ответ: 

$0.63$

## Задание 13

Сколькими способами можно распределить 10 одинаковых яблок между 4 детьми, если каждый ребенок должен получить хотя бы одно яблоко?

Поскольку каждый ребенок должен получить хотя бы одно яблоко, сначала раздадим каждому ребенку по одному яблоку. $10 - 4 = 6$ яблок.

Теперь нужно распределить 6 одинаковых яблок между 4 детьми, при этом дети могут получить от 0 до 6 яблок


Количество способов распределить $n$ одинаковых предметов между $k$ ящиков (в данном случае, 6 яблок между 4 детьми) вычисляется по формуле:

$
\binom{n + k - 1}{k - 1}
$
где $n$ — количество предметов, $k$ — количество ящиков.

В нашем случае $n = 6$ и $k = 4$:
$
\binom{6 + 4 - 1}{4 - 1} = \binom{9}{3}
$


$
\binom{9}{3} = \frac{9!}{3!(9-3)!} = \frac{9!}{3! \cdot 6!} = \frac{9 \times 8 \times 7}{3 \times 2 \times 1} = 84
$

Ответ: 

$ 84 $

## Задание 14

На экзамене студент отвечает на 5 вопросов из билета, который содержит 10 вопросов. Порядок вопросов важен. Сколькими способами студент может выбрать и расположить вопросы для ответа?


Сначала нужно выбрать 5 вопросов из 10. Количество способов выбрать 5 вопросов из 10 без учета порядка вычисляется по формуле сочетаний:

$
\binom{10}{5} = \frac{10!}{5!(10-5)!} = \frac{10!}{5! \cdot 5!} = \frac{10 \times 9 \times 8 \times 7 \times 6}{5 \times 4 \times 3 \times 2 \times 1} = 252
$



После выбора 5 вопросов, их нужно расположить в определенном порядке. Количество способов расположить 5 вопросов равно числу перестановок 5 элементов:

$
5! = 5 \times 4 \times 3 \times 2 \times 1 = 120
$


Общее количество способов выбрать и расположить 5 вопросов из 10 равно произведению числа сочетаний и числа перестановок:

$
\binom{10}{5} \times 5! = 252 \times 120 = 30240
$


Ответ:

$ 30240 $

   $

$
AM = x \sin 60^\circ = x \cdot \frac{\sqrt{3}}{2}

$
DM = x \cos 60^\circ = x \cdot \frac{1}{2}
$

Так как $DM = 7.5$, то:
$
x \cdot \frac{1}{2} = 7.5 \implies x = 7.5 \cdot 2 = 15
$

Теперь мы можем найти периметр трапеции:
$
P = 39 + 2x = 39 + 2 \cdot 15 = 39 + 30 = 69
$
