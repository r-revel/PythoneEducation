Задание 3 (про двигатели)

```python
'''Ваш ответ на задание №3 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''


# Данные
sample_mean1 = 90  # средняя прочность стальных компонентов
sample_mean2 = 88  # средняя прочность алюминиевых компонентов
std_dev1 = 2.3  # стандартное отклонение стальных компонентов
std_dev2 = 2.2  # стандартное отклонение алюминиевых компонентов
sample_size1 = 15  # размер выборки стальных компонентов
sample_size2 = 10  # размер выборки алюминиевых компонентов
alpha = 0.05  # уровень значимости

# Вычисление t-статистики
t_statistic = (sample_mean1 - sample_mean2) / np.sqrt((std_dev1**2 / sample_size1) + (std_dev2**2 / sample_size2))

# Определение критического значения
df = sample_size1 + sample_size2 - 2  # степени свободы
critical_value = stats.t.ppf(1 - alpha / 2, df)

# Вычисление p-значения
p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df))

# Статистический вывод
if np.abs(t_statistic) > critical_value:
    t_test_result = "Отвергаем нулевую гипотезу"
else:
    t_test_result = "Не отвергаем нулевую гипотезу"

if p_value < alpha:
    p_value_result = "Отвергаем нулевую гипотезу"
else:
    p_value_result = "Не отвергаем нулевую гипотезу"

# Вывод результатов
print(f"\nРезультаты теста:")
print(f"t-статистика: {t_statistic:.3f}")
print(f"Критическое значение: {critical_value:.3f}")
print(f"p-значение: {p_value:.4f}")
print(f"Статистический вывод (t-тест): {t_test_result}")
print(f"Статистический вывод (p-значение): {p_value_result}")

# Содержательный вывод
if t_test_result == "Отвергаем нулевую гипотезу" or p_value_result == "Отвергаем нулевую гипотезу":
    content_result = "Есть достаточные доказательства того, что средние прочностные характеристики стальных и алюминиевых компонентов различаются. Следовательно, выбор материала требует дальнейшего анализа."
else:
    content_result = "Нет достаточных доказательств того, что средние прочностные характеристики стальных и алюминиевых компонентов различаются. Следовательно, можно рассмотреть использование алюминиевых компонентов для снижения веса."

print(f"Содержательный вывод: {content_result}")
```

Задание 8 (про фрукты и IQ)

```python
'''Ваш ответ на задание №8 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''


# Данные
oranges = np.array([125, 138, 112, 130, 124, 125, 110, 112, 117, 126])
pears = np.array([124, 102, 142, 130, 131, 124, 127, 115, 108, 109])

# Вычисление дисперсий
var_oranges = np.var(oranges, ddof=1)  # ddof=1 для несмещенного оценщика
var_pears = np.var(pears, ddof=1)

# Выполнение F-теста
f_statistic = var_pears / var_oranges
df1 = len(oranges) - 1  # степени свободы для первой выборки
df2 = len(pears) - 1  # степени свободы для второй выборки

# Критическое значение
alpha = 0.05
critical_value = stats.f.ppf(1 - alpha / 2, df1, df2)

# Вычисление p-значения
p_value = 2 * min(stats.f.cdf(f_statistic, df1, df2), 1 - stats.f.cdf(f_statistic, df1, df2))

# Статистический вывод
if f_statistic > critical_value or f_statistic < 1 / critical_value:
    f_test_result = "Отвергаем нулевую гипотезу"
else:
    f_test_result = "Не отвергаем нулевую гипотезу"

if p_value < alpha:
    p_value_result = "Отвергаем нулевую гипотезу"
else:
    p_value_result = "Не отвергаем нулевую гипотезу"

# Вывод результатов
print(f"F-статистика: {f_statistic:.3f}")
print(f"Критическое значение: {critical_value:.3f}")
print(f"p-значение: {p_value:.4f}")
print(f"Статистический вывод (F-тест): {f_test_result}")
print(f"Статистический вывод (p-значение): {p_value_result}")

# Содержательный вывод
if f_test_result == "Отвергаем нулевую гипотезу" or p_value_result == "Отвергаем нулевую гипотезу":
    content_result = "Есть достаточные доказательства того, что дисперсии обеих групп различаются. Следовательно, диета, состоящая исключительно из апельсинов или груш, влияет на разброс значений IQ."
else:
    content_result = "Нет достаточных доказательств того, что дисперсии обеих групп различаются. Следовательно, диета, состоящая исключительно из апельсинов или груш, не влияет на разброс значений IQ."

print(f"Содержательный вывод: {content_result}")
```

Задание 9 (про самолеты)

```python
'''Ваш ответ на задание №9 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''


# Данные
observed_frequencies = np.array([36, 23, 11, 13, 8, 3, 2, 1, 3])
n = np.sum(observed_frequencies)  # общее количество самолетов

# Оценка параметра p геометрического распределения
mean_value = np.sum(observed_frequencies * np.arange(1, 10)) / n
p = 1 / mean_value

# Вычисление ожидаемых частот
x = np.arange(1, 10)
expected_frequencies = n * stats.geom.pmf(x, p)

# Нужна нормализация ожидаемых частот, иначе будет ошибка 
expected_frequencies *= n / np.sum(expected_frequencies)

# Выполнение χ²-теста
chi2_statistic, p_value = stats.chisquare(observed_frequencies, expected_frequencies)

# Уровень значимости
alpha = 0.05

# Критическое значение
df = len(observed_frequencies) - 1  # степени свободы
critical_value = stats.chi2.ppf(1 - alpha, df)

# Статистический вывод
if chi2_statistic > critical_value:
    chi2_test_result = "Отвергаем нулевую гипотезу"
else:
    chi2_test_result = "Не отвергаем нулевую гипотезу"

if p_value < alpha:
    p_value_result = "Отвергаем нулевую гипотезу"
else:
    p_value_result = "Не отвергаем нулевую гипотезу"

# Вывод результатов
print(f"Оценка λ: {mean_value}")
print(f"χ²-статистика: {chi2_statistic:.3f}")
print(f"Критическое значение: {critical_value:.3f}")
print(f"p-значение: {p_value:.4f}")
print(f"Статистический вывод (χ²-тест): {chi2_test_result}")
print(f"Статистический вывод (p-значение): {p_value_result}")

# Содержательный вывод
if chi2_test_result == "Отвергаем нулевую гипотезу" or p_value_result == "Отвергаем нулевую гипотезу":
    content_result = "Есть достаточные доказательства того, что данные не соответствуют геометрическому распределению. Следовательно, предположение о геометрическом распределении не подтверждается."
else:
    content_result = "Нет достаточных доказательств того, что данные не соответствуют геометрическому распределению. Следовательно, предположение о геометрическом распределении подтверждается."

print(f"Содержательный вывод: {content_result}")

# Построение графика
plt.figure(figsize=(10, 6))
plt.bar(np.arange(1, 10), observed_frequencies, label='Наблюдаемые частоты', alpha=0.7)
plt.bar(np.arange(1, 10), expected_frequencies, label='Ожидаемые частоты', alpha=0.7, color='red')
plt.xlabel('Количество вылетов')
plt.ylabel('Частота')
plt.title('Сравнение наблюдаемых и ожидаемых частот')
plt.legend()
plt.show()
```

Задание 11 (про тигров)

```python
'''Ваш ответ на задание №11
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- в случае, если нулевая гипотеза отвергнута, сделан тест Тьюки для подробного сравнения средних
- сделано дополнение к статистическому выводу по результатам теста Тьюки
- сделан содержательный вывод по поставленной проблеме'''

# Функция для преобразования результатов теста Тьюки в DataFrame
def tukey_hsd_to_dataframe(result):
    # Создаем пустой DataFrame для хранения результатов
    data = {
        'Comparison': [],
        'Statistic': [],
        'p-value': [],
        'Lower CI': [],
        'Upper CI': []
    }
    
    result.confidence_interval(confidence_level=.95)
    
    # Заполняем DataFrame
    for i in range(result.statistic.shape[0]):
        for j in range(result.statistic.shape[0]):
            if i != j:
                comparison = f"({i} - {j})"
                data['Comparison'].append(comparison)
                data['Statistic'].append(result.statistic[i, j])
                data['p-value'].append(result.pvalue[i, j])
                data['Lower CI'].append(result._ci.low[i, j])
                data['Upper CI'].append(result._ci.high[i, j])

    # Преобразуем словарь в DataFrame
    df = pd.DataFrame(data)
    return df

# Данные
masses = {
    'No1': [151, 135, 137, 118, 132, 135, 131, 121, 137, 140, 152, 133, 151, 132],
    'No2': [108, 123, 145, 103, 110, 124, 116, 108, 104, 130, 106, 105, 114, 108],
    'No3': [140, 152, 149, 132, 124, 146, 156, 137, 124, 145, 147, 139, 157, 139],
    'No4': [127, 135, 110, 116, 119, 123, 102, 121, 118, 120, 114, 134, 121, 109]
}

# Преобразование данных в формат для ANOVA
masses_list = [masses[key] for key in masses]
masses_flat = [item for sublist in masses_list for item in sublist]
groups = [key for key in masses for _ in range(len(masses[key]))]

# Выполнение ANOVA
f_statistic, p_value = stats.f_oneway(*masses_list)

# Уровень значимости
alpha = 0.05

# Статистический вывод
if p_value < alpha:
    anova_result = "Отвергаем нулевую гипотезу"
else:
    anova_result = "Не отвергаем нулевую гипотезу"

# Вывод результатов ANOVA
print(f"F-статистика: {f_statistic:.3f}")
print(f"p-значение: {p_value:.4f}")
print(f"Статистический вывод (ANOVA): {anova_result}")

# В случае отвержения нулевой гипотезы, выполнение теста Тьюки
if anova_result == "Отвергаем нулевую гипотезу":
    # Выполнение теста Тьюки
    tukey_results = stats.tukey_hsd(*masses_list)
    # Преобразование результатов в DataFrame для удобства вывода
    tukey_results_df = tukey_hsd_to_dataframe(tukey_results)
    
    tukey_results_df['Reject'] = tukey_results_df['p-value'] < alpha
    # Вывод уникальных значений p-value
    tukey_results_df = tukey_results_df.drop_duplicates(subset=['p-value']).reset_index(drop=True)  
    
    print(tukey_results_df)
    # Визуализация данных
    plt.figure(figsize=(12, 8))

    # График статистики
    plt.errorbar(tukey_results_df['Comparison'], tukey_results_df['Statistic'], 
                yerr=[tukey_results_df['Statistic'] - tukey_results_df['Lower CI'], 
                    tukey_results_df['Upper CI'] - tukey_results_df['Statistic']], 
                fmt='o', capsize=5, label='Статистика с доверительными интервалами')

    # Линия для p-value (обычно p < 0.05 считается значимым)
    plt.axhline(y=0, color='gray', linestyle='--', label='p < 0.05')
  
    # Отметки для отвергнутых гипотез
    for i, reject in enumerate(tukey_results_df['Reject']):
        if reject:
            plt.text(i, tukey_results_df['Statistic'][i], 'Отвергнута', ha='center', va='bottom', color='red')

    # Настройка осей и меток
    plt.xlabel('Сравнение')
    plt.ylabel('Статистика')
    plt.title('Сравнение групп с p-значениями и доверительными интервалами')
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()

    plt.show()
    
    # Дополнение к статистическому выводу
    significant_pairs = tukey_results_df[tukey_results_df['Reject'] == True]['Comparison']
    significant_pairs_str = ', '.join(significant_pairs)
    
    additional_result = f"Существенное отличие в средних наблюдается в следующих парах групп: {significant_pairs_str}"
else:
    additional_result = "Не требуется дополнительное сравнение, так как нулевая гипотеза не отвергнута."

# Содержательный вывод
if anova_result == "Отвергаем нулевую гипотезу":
    content_result = "Есть достаточные доказательства того, что тип корма влияет на среднюю массу тигров. Следовательно, разные типы корма приводят к различиям в массе тигров."
else:
    content_result = "Нет достаточных доказательств того, что тип корма влияет на среднюю массу тигров. Следовательно, разные типы корма не приводят к значимым различиям в массе тигров."

print(f"Дополнение к статистическому выводу: {additional_result}")
print(f"Содержательный вывод: {content_result}")
```
